```python
import pandas as pd
```

```python
# Read in the data
mbti_table = pd.read_csv('MBTI.csv')
mbti_table.head()
```

```python
srq_table = pd.read_csv('SRQ.csv')
srq_table.head()
```

```python
vark_table = pd.read_csv('VARK.csv')
vark_table.head()
```

```python
# Join the tables on name
combined_table = mbti_table.merge(srq_table, on='Name').merge(vark_table, on='Name')
combined_table.head()
```

```python
# Save the combined table
combined_table.to_csv('combined_table.csv', index=False)
```
